一、基础创建与查看

```python
#1.导入 pandas，用字典创建DataFrame
import pandas as pd
data1= {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data1)
df1
```

```python
#2.查看上述 DataFrame 的前 1 行、形状、列名、数据类型。
print("\n前1行数据:",df1.head(1))
print("\n数据形状(行数,列数):", df1.shape)
print("\n列名:", df1.columns.tolist())
print("\n数据类型:",df1.dtypes)
```

```python
# 3. 用列表创建DataFrame
data2 = [[1, "A"], [2, "B"], [3, "C"]]
df2 = pd.DataFrame(data2, columns=["id", "tag"])
df2
```

二、列与行操作

```python
# 1. 选取 score 和 gender 两列
df1[['score', 'gender']]
```

```python
# 2. 选取索引 0、2 的行
df1.loc[[0, 2]]
```

```python
# 3. 筛选 score>80 的行
df1[df1["score"] > 80]
```

```python
# 4. 新增grade列：score≥90为优秀，否则为良好
df1["grade"] = df1["score"].apply(lambda x: "优秀" if x >= 90 else "良好")
print("\n4. 新增grade列后：")
print(df1)
```

```python

# 5. 删除gender列
df1_drop = df1.drop(columns="gender")
print("\n5. 删除gender列后：")
print(df1_drop)
```

三、缺失值

```python
# 1. 创建含缺失值的DataFrame，检测缺失值并统计每列缺失数量
data = {"x":[1, None, 3], "y":[None, 5, 6]}
df = pd.DataFrame(data)
print("原始含缺失值的DataFrame:")
print(df)

# 检测缺失值
print("\n缺失值布尔检测:")
print(df.isnull())

# 统计每列缺失数量
print("\n每列缺失值数量:")
print(df.isnull().sum())
```

```python
# 2. 用0填充缺失值；删除含缺失值的行
df_fill = df.fillna(0)
print("\n用0填充缺失值后：")
print(df_fill)

df_drop = df.dropna()
print("\n删除含缺失值的行后：")
print(df_drop)
```

四、分组聚合

```python
# 1. 按 gender 分组，求 score 的均值和最大值
result1 = df1.groupby('gender')['score'].agg(['mean', 'max'])
print("按 gender 分组的 score 均值和最大值：")
print(result1)
```

```python
# 2. 按 gender 分组，统计每组人数
result2 = df1.groupby('gender').size()
print("\n按 gender 分组的每组人数：")
print(result2)
```

五、排序与去重

```python
# 1. 按 score 升序排序
df_sorted = df1.sort_values(by='score', ascending=True)
print("按 score 升序排序后的 DataFrame：")
print(df_sorted)
```

```python
# 2.添加重复数据（Tom,85,男）
new_row = pd.DataFrame([{"name": "Tom", "score": 85, "gender": "男"}])
df = pd.concat([df1, new_row], ignore_index=True)
print("\n添加重复数据后:")
print(df)

# 去重（默认保留第一次出现的行）
df_dedup = df.drop_duplicates(keep="first")
print("\n去重后:")
print(df_dedup)
```

六、合并与拼接

```python
# 1. 横向合并（按id内连接）
df1 = pd.DataFrame({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame({"id":[1,2], "age":[20,30]})
df_merge = pd.merge(df1, df2, on="id")
print("1. 横向合并结果：")
print(df_merge)
```

```python
# 2. 纵向拼接并重置索引
df_a = pd.DataFrame({"val":[1,2]})
df_b = pd.DataFrame({"val":[3,4]})
df_concat = pd.concat([df_a, df_b], ignore_index=True)
print("\n2. 纵向拼接并重置索引结果：")
print(df_concat)
```

七、时间处理

```python
# 1. 创建时间列，转为datetime格式
date_list = ["2025-01-01","2025-01-02","2025-01-03"]
df = pd.DataFrame({"date": date_list})
df["date"] = pd.to_datetime(df["date"])
print("1. 转换为datetime格式：")
print(df)
```

```python
# 2. 提取月份、星期
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.day_name()   # 英文星期
# df["weekday_cn"] = df["date"].dt.dayofweek+1  # 数字星期(1=周一,7=周日)
print("\n2. 提取月份、星期：")
print(df)
```

八、高级操作

```python
data1 = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78], 
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data1)

# 1. 对 score 列做标准化
score_mean = df1['score'].mean()
score_std = df1['score'].std(ddof=0)
df1['score_standardized'] = (df1['score'] - score_mean) / score_std

print("\n标准化后的数据:")
print(df1)
```

```python
# 2. 透视表：按 gender 分组，计算 score 中位数
pivot_result = df1.pivot_table(
    index='gender',
    values='score',
    aggfunc='median'
)

print("\n按 gender 分组的 score 中位数：")
print(pivot_result)
```
