```python
import pandas as pd

#1.
data_dict = {
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
}
df1 = pd.DataFrame(data_dict)
print("字典创建的DataFrame:")
print(df1)

#2.
print("\n前1行数据:")
print(df1.head(1))

print("\n数据形状(行数,列数):", df1.shape)

print("\n列名:", df1.columns.tolist())

print("\n数据类型:")
print(df1.dtypes)

# 3. 用列表创建DataFrame，列名设为id、tag
data_list = [[1, "A"], [2, "B"], [3, "C"]]
df2 = pd.DataFrame(data_list, columns=["id", "tag"])
print("\n列表创建的DataFrame:")
print(df2)
```

```python
# 1. 选取 score 和 gender 两列
df_col = df1[["score", "gender"]]
print("1. 选取score和gender列：")
print(df_col)

# 2. 选取索引0、2的行
df_row = df1.loc[[0, 2]]
print("\n2. 选取索引0、2的行：")
print(df_row)

# 3. 筛选 score>80 的行
df_filter = df1[df1["score"] > 80]
print("\n3. 筛选score>80的行：")
print(df_filter)

# 4. 新增grade列：score≥90为优秀，否则为良好
df1["grade"] = df1["score"].apply(lambda x: "优秀" if x >= 90 else "良好")
print("\n4. 新增grade列后：")
print(df1)

# 5. 删除gender列
df1_drop = df1.drop(columns="gender")
print("\n5. 删除gender列后：")
print(df1_drop)
```

```python
# 1. 创建含缺失值的DataFrame，检测缺失值并统计每列缺失数量
data = {"x":[1, None, 3], "y":[None, 5, 6]}
df = pd.DataFrame(data)
print("原始含缺失值的DataFrame：")
print(df)

# 检测缺失值
print("\n缺失值布尔检测：")
print(df.isnull())

# 统计每列缺失数量
print("\n每列缺失值数量：")
print(df.isnull().sum())

# 2. 用0填充缺失值；删除含缺失值的行
df_fill = df.fillna(0)
print("\n用0填充缺失值后：")
print(df_fill)

df_drop = df.dropna()
print("\n删除含缺失值的行后：")
print(df_drop)
```

```python
# 1. 按 gender 分组，求 score 的均值和最大值
group_agg = df.groupby("gender")["score"].agg(["mean", "max"])
print("1. 按性别分组，分数均值、最大值：")
print(group_agg)

# 2. 按 gender 分组，统计每组人数
group_count = df.groupby("gender").size()
# 也可以用 count()：df.groupby("gender")["name"].count()
print("\n2. 按性别分组，每组人数：")
print(group_count)
```

```python
# 1. 按score升序排序
df_sort = df.sort_values(by="score", ascending=True)
print("1. 按score升序排序：")
print(df_sort)

# 2. 添加重复行并去重
# 追加Tom的重复数据
df_new = pd.concat([df, pd.DataFrame([["Tom", 85, "男"]], columns=df.columns)], ignore_index=True)
print("\n添加重复行后：")
print(df_new)

# 去重
df_drop_dup = df_new.drop_duplicates()
print("\n去重后：")
print(df_drop_dup)
```

```python
# 1. 横向合并（按id内连接）
df1 = pd.DataFrame({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame({"id":[1,2], "age":[20,30]})
df_merge = pd.merge(df1, df2, on="id")
print("1. 横向合并结果：")
print(df_merge)

# 2. 纵向拼接并重置索引
df_a = pd.DataFrame({"val":[1,2]})
df_b = pd.DataFrame({"val":[3,4]})
df_concat = pd.concat([df_a, df_b], ignore_index=True)
print("\n2. 纵向拼接并重置索引结果：")
print(df_concat)
```

```python
# 1. 创建时间列，转为datetime格式
date_list = ["2025-01-01","2025-01-02","2025-01-03"]
df = pd.DataFrame({"date": date_list})
df["date"] = pd.to_datetime(df["date"])
print("1. 转换为datetime格式：")
print(df)

# 2. 提取月份、星期
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.day_name()   # 英文星期
# df["weekday_cn"] = df["date"].dt.dayofweek+1  # 数字星期(1=周一,7=周日)
print("\n2. 提取月份、星期：")
print(df)
```

```python
import numpy as np

# 1. 对score列做标准化：(值-均值)/标准差
df["score_standard"] = (df["score"] - df["score"].mean()) / df["score"].std()
print("1. score标准化结果：")
print(df)

# 2. 透视表：行为gender，值为score，聚合函数为中位数
pivot_table = pd.pivot_table(df, index="gender", values="score", aggfunc="median")
print("\n2. 透视表（性别分组，分数中位数）：")
print(pivot_table)
```
